```python
import pandas as pd
import os

df = pd.read_csv("../data/fraud_transactions.csv")
print(f"Original rows: {len(df):,}")
print(f"Original file size: {os.path.getsize('../data/fraud_transactions.csv') / (1024*1024):.2f} MB")
```

```python
df_small = df.sample(frac=0.2, random_state=42)

output_path = "../data/fraud_transactions_small.csv"
df_small.to_csv(output_path, index=False)

size_mb = os.path.getsize(output_path) / (1024 * 1024)
print(f"Sampled file size: {size_mb:.2f} MB")
print(f"Sampled rows: {len(df_small):,}")
```
